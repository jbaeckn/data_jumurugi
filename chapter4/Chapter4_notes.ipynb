```python
import pandas as pd
from glob import glob
```

```python
glob('/Users/jonghobaeck/downloads/지역*.xls')
```

```python
stations_files = glob('/Users/jonghobaeck/downloads/지역*.xls')
stations_files
```

```python
tmp_raw = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
```

```python
station_raw = pd.concat(tmp_raw)
station_raw
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'Oil_store':station_raw['상호'],
                        '주소':station_raw['주소'],
                        '가격':station_raw['휘발유'],
                        '셀프':station_raw['셀프여부'],
                        '상표':station_raw['상표']})

stations
```

```python
stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```

```python
stations['구'].unique()
```

```python
stations[stations['구']=='서울특별시']
```

```python
stations.loc[stations['구']=='서울특별시', '구'] = '성동구'
stations['구'].unique()
```

```python
stations.loc[stations['구']=='특별시', '구'] = '도봉구'
stations['구'].unique()
```

```python
stations[stations['가격']=='-']
```

```python
stations = stations[stations['가격'] != '-']
stations.head()
```

```python
stations['가격'] = [float(value) for value in stations['가격']]
```

```python
stations.reset_index(inplace=True)
```

```python
del stations['index']
```

```python
stations
```
